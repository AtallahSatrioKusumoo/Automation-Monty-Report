```python
ascii_art = r"""
  /$$$$$$    /$$               /$$ /$$           /$$             /$$$$$$$                                               /$$
 /$$__  $$  | $$              | $$| $$          | $$            | $$__  $$                                             | $$
| $$  \ $$ /$$$$$$    /$$$$$$ | $$| $$  /$$$$$$ | $$$$$$$       | $$  \ $$ /$$$$$$   /$$$$$$  /$$  /$$$$$$   /$$$$$$$ /$$$$$$
| $$$$$$$$|_  $$_/   |____  $$| $$| $$ |____  $$| $$__  $$      | $$$$$$$//$$__  $$ /$$__  $$|__/ /$$__  $$ /$$_____/|_  $$_/
| $$__  $$  | $$      /$$$$$$$| $$| $$  /$$$$$$$| $$  \ $$      | $$____/| $$  \__/| $$  \ $$ /$$| $$$$$$$$| $$        | $$
| $$  | $$  | $$ /$$ /$$__  $$| $$| $$ /$$__  $$| $$  | $$      | $$     | $$      | $$  | $$| $$| $$_____/| $$        | $$ /$$
| $$  | $$  |  $$$$/|  $$$$$$$| $$| $$|  $$$$$$$| $$  | $$      | $$     | $$      |  $$$$$$/| $$|  $$$$$$$|  $$$$$$$  |  $$$$/
|__/  |__/   \___/   \_______/|__/|__/ \_______/|__/  |__/      |__/     |__/       \______/ | $$ \_______/ \_______/   \___/
                                                                                        /$$  | $$
                                                                                       |  $$$$$$/
                                                                                        \______/
"""


frame_width = 160
frame_height = 20


art_lines = ascii_art.strip("\n").split("\n")
padding_top = (frame_height - len(art_lines)) // 2
padding_bottom = frame_height - len(art_lines) - padding_top


print("=" * frame_width)
for _ in range(padding_top):
    print("|" + " " * (frame_width - 2) + "|")
for line in art_lines:
    line_content = line.center(frame_width - 2)
    print(f"|{line_content}|")
for _ in range(padding_bottom):
    print("|" + " " * (frame_width - 2) + "|")
print("=" * frame_width)
```
